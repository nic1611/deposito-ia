# Dimensionality Reduction

```python
from keras.datasets import mnist
import matplotlib.pyplot as plt
```

```python
(X,y),(_,_) = mnist.load_data()
print(X.shape, y.shape)
```

```python
X = X.reshape(len(X), -1)
X.shape
```

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from umap import UMAP
```

## PCA

```python
x_pca = PCA(n_components=2).fit_transform(X)
plt.figure(figsize=(10,10))
plt.scatter(x_pca[:,0], x_pca[:,1], c=y)
plt.show()
```

## LDA

```python
x_lda = LDA(n_components=2).fit_transform(X, y)
print(x_lda.shape)
plt.figure(figsize=(10,10))
plt.scatter(x_lda[:,0], x_lda[:,1], c=y)
plt.show()
```

## t-SNE

```python
x_tsne = TSNE().fit_transform(X[:1000,:])
print(x_tsne.shape)
plt.figure(figsize=(10,10))
plt.scatter(x_tsne[:1000,0], x_tsne[:1000,1], c=y[:1000])
plt.show()
```

## UMAP

```python
x_umap = UMAP(n_neighbors=10, min_dist=0.1, metric='correlation').fit_transform(X[:1000,:])
print(x_umap.shape)
plt.figure(figsize=(10,10))
plt.scatter(x_umap[:1000,0], x_umap[:1000,1], c=y[:1000])
plt.show()
```
